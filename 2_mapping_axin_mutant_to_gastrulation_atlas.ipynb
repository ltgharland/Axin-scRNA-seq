## Perform rPCA Integration of the Original Gastrulation Atlas including E6.5 --> E8.5

```r
# Subset the downsampled reference to only include the cells from the original 2019 gastrulation atlas
reference_original <- reference[,reference@meta.data$embryo_version %in% "Original"]

# set the features to intgerate as all the gene names
features_to_integrate <- row.names(reference_original)

# Define the stages you want to integrate
stages_to_integrate <- c(
    "E6.5_Original",
    "E6.75_Original",
    "E7.0_Original", 
    "E7.25_Original", 
    "E7.5_Original", 
    "E7.75_Original",
    "E8.0_Original",
    "E8.25_Original",
    "E8.5_Original"
)

# Define a function to process and integrate each stage
integrate_stage <- function(seurat_object, stage, split.by = "sample", num_features = 2000, k_anchor = 5, k_weight = 100) {
  message(stage)
  
  # Subset the data for the current stage
  object <- seurat_object[, seurat_object@meta.data$stage_updated == stage]
  
  # Split object based on the specified split.by parameter
  object.split <- SplitObject(object, split.by = split.by)
  
  # Normalize and identify variable features for each dataset independently
  object.split <- lapply(X = object.split, FUN = function(x) {
    x <- FindVariableFeatures(x, selection.method = "vst", nfeatures = num_features)
    return(x)
  })
  
  features <- SelectIntegrationFeatures(object.list = object.split, nfeatures = num_features)
  
  # Select features that are repeatedly variable across datasets for integration
  object.split <- lapply(X = object.split, FUN = function(x) {
    x <- ScaleData(x, features = features, verbose = FALSE)
    x <- RunPCA(x, features = features, verbose = FALSE)
    return(x)
  })
  
  # Find Integration Anchors
  if (length(unique(object@meta.data[[split.by]])) > 1) {   
    rPCA_anchors <- FindIntegrationAnchors(object.list = object.split, 
                                           anchor.features = features, 
                                           reduction = "rpca", 
                                           k.anchor = k_anchor,
                                           scale = FALSE)
    
    # Integrate the data
    object_integrated <- IntegrateData(anchorset = rPCA_anchors,
                                       k.weight = k_weight,
                                       features.to.integrate = rownames(seurat_object),
                                       new.assay.name = "within_timepoint_integration")
    # Return the integrated object
    return(object_integrated)
  } else {
    return(NULL)
  }
}

# Set parameters
num_features <- 2000
k_anchor <- 5
k_weight <- 100

# Apply the function to each stage
integrated_objects <- lapply(stages_to_integrate, function(stage) {
  integrate_stage(seurat_object = reference_original, stage = stage, split.by = "sample", num_features = num_features, k_anchor = k_anchor, k_weight = k_weight)
})

# Add the names to the integrated object list
stages_integrated <- c(
    "E6.5_Original",
    "E7.0_Original", 
    "E7.25_Original", 
    "E7.5_Original", 
    "E7.75_Original",
    "E8.0_Original",
    "E8.25_Original",
    "E8.5_Original"
)

names(integrated_objects) <- stages_integrated

# Add the E6.75 gastrula dataset that only has one sample
E6.75_Original <- reference_original[, reference_original@meta.data$stage_updated == "E6.75_Original"]
E6.75_Original <- FindVariableFeatures(E6.75_Original, selection.method = "vst", nfeatures = num_features)
E6.75_Original <- ScaleData(E6.75_Original, verbose = FALSE)
E6.75_Original <- RunPCA(E6.75_Original, verbose = FALSE)

integrated_objects$E6.75_Original <- E6.75_Original
```

```r
### WRITE A FUNCTION TO: extract VariableFeatures from a list of Seurat objects

extract_variable_features <- function(object_list) {
  variable_features_list <- lapply(object_list, function(seurat_object) {
    VariableFeatures(seurat_object)
  })
  return(variable_features_list)
}

## USE THE FUNCTION: Assuming integrated_objects is your list of Seurat objects
all_variable_features <- extract_variable_features(integrated_objects)

# Combine all VariableFeatures into a single vector and make it unique
combined_variable_features <- unique(unlist(all_variable_features))
str(combined_variable_features)

## USE THE OVERLAPPING IDENTIFIED VARIABLE FEATURES TO PERFORM PCA ON ALL THE DATATSETS ACROSS TIMEPOINTS
overlapping_features <- combined_variable_features


## SAVE THE VARIABLE FEATURES ###
saveRDS(all_variable_features, "projects/10_Axin_1_2/github_submission/outputs/2/pipeline_2_all_variable_features_E85_original_atlas.rds")
```

```r
## Write a Function to Perform Seurat Analyses

run_seurat_analysis <- function(seurat_object, dims_to_use = 1:30, cluster_resolution = 2) {
  
  # Run UMAP, Find Neighbors, and Find Clusters
  seurat_object <- FindNeighbors(seurat_object, reduction = "pca", dims = dims_to_use) %>% FindClusters(., resolution = cluster_resolution)
  
  return(seurat_object)
}

object_integrated <- run_seurat_analysis(object_integrated)
```
